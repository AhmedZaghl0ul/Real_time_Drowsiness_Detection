```python
import cv2
import dlib
import tensorflow
import numpy as np
import mediapipe as mp
from tensorflow.keras.models import load_model
import time
import matplotlib.pyplot as plt
```

```python
video_path = r"preview.mp4"

model = load_model("mobilenet_eye_model.h5")

mp_face_mesh = mp.solutions.face_mesh
face_mesh = mp_face_mesh.FaceMesh(static_image_mode=False,
                                  max_num_faces=1,
                                  refine_landmarks=True,
                                  min_detection_confidence=0.5,
                                  min_tracking_confidence=0.5)

cap = cv2.VideoCapture(0)
prev_time = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    curr_time = time.time()
    fps = 1 / (curr_time - prev_time)
    prev_time = curr_time
    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = face_mesh.process(rgb)

    if results.multi_face_landmarks:
        face = results.multi_face_landmarks[0]

        left_eye_idx = [33, 160, 158, 133, 153, 144]
        right_eye_idx = [362, 385, 387, 263, 373, 380]

        h, w, _ = frame.shape

        left_points = [(int(face.landmark[i].x * w), int(face.landmark[i].y * h)) for i in left_eye_idx]
        right_points = [(int(face.landmark[i].x * w), int(face.landmark[i].y * h)) for i in right_eye_idx]

        for p in left_points + right_points:
            cv2.circle(frame, p, 2, (0, 255, 0), -1)

        resized = cv2.resize(frame, (128, 128))
        inp = np.expand_dims(resized / 255.0, axis=0)

        pred = model.predict(inp)[0][0]

        label = "Drowsy" if pred > 0.5 else "Awake"
        color = (0, 0, 255) if label == "Drowsy" else (0, 255, 0)

        cv2.putText(frame, f"{label}: {pred:.2f}", (30, 50),
                    cv2.FONT_HERSHEY_SIMPLEX, 1.2, color, 2)
        
    cv2.putText(frame, f"FPS: {int(fps)}", (30, 90), 
                cv2.FONT_HERSHEY_SIMPLEX, 1.2, (255, 0, 0), 2)
    cv2.imshow("Drowsiness Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
